> ## Advantages of DBSCAN:
- Is great at seperating clusters of high density versus clusters of low density within a given dataset
- Works great with noisy data. Handles outliers efficiently within the dataset

> ## Disadvantages of DBSCAN:
- Does not work well when dealing with clusters of varying densities. While DBSCAN is great at separating high density clusters from low sendity clusters, DBSCAN striggles with clusters of similar density
- Struggles with high dimensionality data. I know, this entire article I have stated how DBSCAN is great at contorting the data into different dimensions and shapes. However, DBSCAN can only go so far, if given data with too many dimensions, DBSCAN suffers

## Importing Required Libraries


```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

## Importing the Dataset

```python
dataset = pd.read_csv('./Mall_Customers.csv')
```

## Exploring the dataset

```python
dataset.head()
```

> ### Considering only `Annual Income` and `Spending Score` features

```python
X = dataset.iloc[:, [3, 4]].values
```

> Applying DBSCAN algorithm

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=3, min_samples=4)
```

> Fitting the model

```python
model = dbscan.fit(X)
labels = model.labels_
```

```python
from sklearn import metrics
```

> Identifying the points which makes up our core points

```python
sample_cores = np.zeros_like(labels, dtype=bool)
sample_cores[dbscan.core_sample_indices_] = True
```

> Calculating the number of clusters

```python
n_clusters = len(set(labels))- (1 if -1 in labels else 0)
```

```python
n_clusters
```

> Calculating Silhouette Score

```python
print(metrics.silhouette_score(X, labels))
```
